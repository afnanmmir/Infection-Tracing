# Import Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# import .gexf file
G = nx.read_gexf("2020-07-01__to__2020-07-05__15-overlap__hashmap.gexf")
# graph its degree distribution as a normal line plot and log log plot
# degree distribution
degrees = [G.degree(n) for n in G.nodes()]
degrees.sort(reverse=True)
plt.plot(degrees)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Count")
plt.savefig("../docs/M2/imgs/5_day_degree_distribution.png")
plt.show()
# log log degree distribution
plt.loglog(degrees)
plt.title("Log Log Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Count")
plt.savefig("../docs/M2/imgs/5_day_log_log_degree_distribution.png")
plt.show()
```
